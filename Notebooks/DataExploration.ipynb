```python
first_file = files_df.iloc[0]['full_path']
df_first   = pd.read_csv(first_file)

timestamps = df_first['timestamp']
time_diffs = timestamps.diff().dropna()

avg_time_diff = time_diffs.mean()

# Auto-detect unit
if avg_time_diff > 1000:
    frequency_hz = 1_000_000 / avg_time_diff
    unit = "microseconds"
elif avg_time_diff > 1:
    frequency_hz = 1000 / avg_time_diff
    unit = "milliseconds"
else:
    frequency_hz = 1 / avg_time_diff
    unit = "seconds"

# ── Diagnostics ────────────────────────────────────────────────────────────────

# Duplicates
n_duplicates = timestamps.duplicated().sum()

# Negative jumps (out of order)
n_negative = (time_diffs < 0).sum()

# Gaps (time diff significantly larger than average)
gap_threshold = avg_time_diff * 3
n_gaps        = (time_diffs > gap_threshold).sum()
max_gap       = time_diffs.max()

print(f"Total samples:       {len(df_first)}")
print(f"Detected unit:       {unit}")
print(f"Sampling frequency:  {frequency_hz:.2f} Hz")
print(f"Avg time diff:       {avg_time_diff:.4f} {unit}")
print(f"Std time diff:       {time_diffs.std():.4f} {unit}  (0 = perfectly regular)")
print(f"Duplicate timestamps:{n_duplicates}")
print(f"Negative jumps:      {n_negative}")
print(f"Gaps (>{gap_threshold:.1f}):    {n_gaps}  (max gap: {max_gap:.2f})")
```
