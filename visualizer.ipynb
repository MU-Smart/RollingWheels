```python
import pandas as pd
import matplotlib.pyplot as plt
import folium
```

# Dresden Data

```python

# Load the CSV file
df = pd.read_csv('./Dresden-July15-18/Phone9/export1531922076/SDC_3657e985a2152942SensorGPSData.csv')

# Display first few rows to understand the structure
print(df.head())

# Plot GPS data (assuming latitude and longitude columns)
# Create a map centered on the mean of the coordinates
map_center = [df['attr_lat'].mean(), df['attr_lng'].mean()]
m = folium.Map(location=map_center, zoom_start=12)

# Add markers for each GPS point
for idx, row in df.iterrows():
    folium.CircleMarker(
        location=[row['attr_lat'], row['attr_lng']],
        radius=3,
        popup=f"Time: {row['attr_time']}",
        color='blue',
        fill=True,
        fillColor='blue',
        fillOpacity=0.6
    ).add_to(m)

# Display the map
m.save('Dresden-July15-18.html')
```

```python
# Load the accelerometer data CSV
accel_df = pd.read_csv('./Dresden-July15-18/Phone9/export1531922076/SDC_3657e985a2152942SensorAccelerometerData.csv')

# Display first few rows to check column names
print(accel_df.head())

# Plot x, y, z acceleration over time
plt.figure(figsize=(15, 6))
plt.plot(accel_df['attr_time'], accel_df['attr_x'], label='X')
plt.plot(accel_df['attr_time'], accel_df['attr_y'], label='Y')
plt.plot(accel_df['attr_time'], accel_df['attr_z'], label='Z')
plt.xlabel('Time')
plt.ylabel('Acceleration (m/s²)')
plt.title('Accelerometer Data (X, Y, Z) Over Time')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# Load the gyroscope data CSV
gyro_df = pd.read_csv('./Dresden-July15-18/Phone9/export1531922076/SDC_3657e985a2152942SensorGyroscopeData.csv')

# Display first few rows to check column names
print(gyro_df.head())

# Plot x, y, z gyroscope data over time
plt.figure(figsize=(15, 6))
plt.plot(gyro_df['attr_time'], gyro_df['attr_x'], label='X')
plt.plot(gyro_df['attr_time'], gyro_df['attr_y'], label='Y')
plt.plot(gyro_df['attr_time'], gyro_df['attr_z'], label='Z')
plt.xlabel('Time')
plt.ylabel('Gyroscope (rad/s)')
plt.title('Gyroscope Data (X, Y, Z) Over Time')
plt.legend()
plt.tight_layout()
plt.show()
```
